### Thông tin chung về dataset
Tập dataset được em kết hợp trong bài tập này được lấy từ các nguồn sau:
- https://data.humdata.org/dataset/novel-coronavirus-2019-ncov-cases?
- https://ourworldindata.org/covid-cases

Em sẽ dùng python để tiền xử lý dữ liệu và sau đó dùng PowerBI để trực quan hoá

```python
# import các thư viện cần thiết
import numpy as np
import pandas as pd
```

```python
# Đọc các file csv 
confirmed_df = pd.read_csv('time_series_covid19_confirmed_global.csv')
deaths_df = pd.read_csv('time_series_covid19_deaths_global.csv')
recovered_df = pd.read_csv('time_series_covid19_recovered_global.csv')
```

```python
# Loại bỏ các cột không cần thiết
confirmed_df.drop(['Province/State', 'Lat', 'Long', ], axis=1, inplace=True)
deaths_df.drop(['Province/State', 'Lat', 'Long', ], axis=1, inplace=True)
recovered_df.drop(['Province/State', 'Lat', 'Long', ], axis=1, inplace=True)
```

```python
# Hiển thị 5 dòng đầu cù dataframe Confirm_df
confirmed_df.head()
```

```python
# Vì trong tập dữ liệu này có bao gồm các dòng dữ liệu của từng thành phố trong 1 khu vực
# Nên em sẽ Group by theo trường Country/Region để tính tổng số trường hợp các ca cho từng khu vực
confirmed_df = confirmed_df.groupby('Country/Region').sum().reset_index()
deaths_df = deaths_df.groupby('Country/Region').sum().reset_index()
recovered_df = recovered_df.groupby('Country/Region').sum().reset_index()
```

```python
# Các ca nhiễm
confirmed_df.head()
```

```python
# Các ca tử vong
deaths_df.head()
```

```python
# Các ca hồi phục
recovered_df.head()
```

```python
# Chuyển từ dạng wide format sang wide format
confirmed_df = confirmed_df.melt(id_vars='Country/Region', var_name='Date', value_name='Confirmed')
deaths_df = deaths_df.melt(id_vars='Country/Region', var_name='Date', value_name='Death')
recovered_df = recovered_df.melt(id_vars='Country/Region', var_name='Date', value_name='Recovered')
```

```python
confirmed_df.head()
```

```python
deaths_df.head()
```

```python
recovered_df.head()
```

```python
# Gộp các dataframe lại theo trường 'Country/Region' và 'Date'
df_cd = pd.merge(confirmed_df, deaths_df, on=['Country/Region', 'Date'])
df_cdr = pd.merge(df_cd, recovered_df, on = ['Country/Region', 'Date'])
```

```python
df_cdr
```

```python
# Xuất ra file csv mới để tiến hành trực quan hoá dữ liệu bằng PowerBI
df_cdr.to_csv('data_covid_19.csv', index=False)
```
